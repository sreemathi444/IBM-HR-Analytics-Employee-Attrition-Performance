```python
# to find the arrtition by age group

# Define age groups
bins = [18, 20, 30, 40, 50, 60]
labels = ["18-20", "21-30", "31-40", "41-50", "51+"]

# Create age group column
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

# Count attrition cases per age group
attrition_by_age = df[df["Attrition"] == "Yes"]["Age Group"].value_counts()

# Print result
print(attrition_by_age)
```

```python
df[df["Attrition"] == "Yes"]["Age Group"].value_counts(normalize= True)*100
```

```python
# to find arrition by salary eange

# Convert 'Attrition' column to binary (1 for 'Yes', 0 for 'No')
df['Attrition_Flag'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Define salary bins
bins = [0, 3000, 6000, 9000, 12000, 15000, df['MonthlyIncome'].max()]
labels = ["<3K", "3K-6K", "6K-9K", "9K-12K", "12K-15K", ">15K"]

# Create a new column for salary ranges
df['Salary_Range'] = pd.cut(df['MonthlyIncome'], bins=bins, labels=labels, include_lowest=True)

# Calculate attrition rate by salary range
attrition_by_salary = df[df["Attrition"] == "Yes"]["Salary_Range"].value_counts(normalize = True) *100

# Display result
print(attrition_by_salary)
```
